<b> Chaos Check </b>

```python
import sim
```

```python
particle_chaos_1 = sim.Particle(
    position=[2.0, 0.0, 0.0],
    velocity=[0.0, 1500.0, 300.0],
    mass=1.0,
    charge=-1.0
)
particle_chaos_2 = sim.Particle(
    position=[2.0 + 1e-6, 0.0, 0.0],  # Slightly perturbed initial position
    velocity=[0.0, 1500.0, 300.0],
    mass=1.0,
    charge=-1.0
)
sim.calculateDistanceOverTime(particle_chaos_1, particle_chaos_2, steps = 100000)
```

```python
sim.plot_double_trajectory(particle_chaos_1, particle_chaos_2)
```

```python
# Particle A: The Baseline
particle_A = sim.Particle(
    position=[2.0, 0.0, 0.0], 
    velocity=[0.0, 1500.0, 200.0], 
    mass=1.0, 
    charge=-1.0
)

# Particle B: The "Butterfly Effect" Test
# Note the tiny 0.001 shift in the starting X position
particle_B = sim.Particle(
    position=[2.001, 0.0, 0.0], 
    velocity=[0.0, 1500.0, 200.0], 
    mass=1.0, 
    charge=-1.0
)
sim.calculateDistanceOverTime(particle_A, particle_B, steps = 100000)
```

```python
sim.plot_double_trajectory(particle_A, particle_B)
```
