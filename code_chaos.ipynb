<b> Chaos Check </b>

Take two close points ${10}^{-6}$ apart.
Show that they are chaotic by calculating distance over time between two particles.
Chaos

```python
import sim
```

```python
chaos_particle_A = sim.Particle(position=[1.5, 0.0, 0.5], velocity=[0.0, 0.0, 50.0], mass=1.0, charge=-1.0)
chaos_particle_B = sim.Particle(position=[1.500001, 0.0, 0.5], velocity=[0.0, 0.0, 50.0], mass=1.0, charge=-1.0)
```

```python
sim.calculateDistanceOverTime(chaos_particle_A, chaos_particle_B, steps = 500000)
```

```python
sim.plot_double_trajectory(chaos_particle_A, chaos_particle_B, steps = 500000)
```

```python
scattering_particle_1 = sim.Particle(
    position=[-5.0, 1.0, 0.5], 
    velocity=[3500.0, 0.0, 0.0], # Fast approach from the left
    mass=1.0, 
    charge=-1.0
)
scattering_particle = sim.Particle(
    position=[-5.0, 1.0, 0.5], 
    velocity=[3500.0 + 1e-6, 0.0, 0.0], # Fast approach from the left
    mass=1.0, 
    charge=-1.0
)
sim.calculateDistanceOverTime(scattering_particle_1, scattering_particle, steps = 500000)
```

```python
sim.plot_double_trajectory(scattering_particle_1, scattering_particle, steps = 500000)
```
