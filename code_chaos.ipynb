<b> Chaos Check </b>

Take two close points ${10}^{-6}$ apart.
Show that they are chaotic by calculating distance over time between two particles.
Chaos

```python
import sim
```

```python
# Particle A: The Baseline
particle_A = sim.Particle(
    position=[2.0, 0.0, 0.0], 
    velocity=[0.0, 1500.0, 200.0], 
    mass=1.0, 
    charge=-1.0
)

# Particle B: The "Butterfly Effect" Test
# Note the tiny 0.001 shift in the starting X position
particle_B = sim.Particle(
    position=[2.001, 0.0, 0.0], 
    velocity=[0.0, 1500.0, 200.0], 
    mass=1.0, 
    charge=-1.0
)
sim.calculateDistanceOverTime(particle_A, particle_B, steps = 750000)
```

```python
sim.plot_double_trajectory(particle_A, particle_B, steps = 750000)
```

```python
chaos_particle_A = sim.Particle(position=[1.5, 0.0, 0.5], velocity=[0.0, 0.0, 50.0], mass=1.0, charge=-1.0)
chaos_particle_B = sim.Particle(position=[1.500001, 0.0, 0.5], velocity=[0.0, 0.0, 50.0], mass=1.0, charge=-1.0)
```

```python
sim.calculateDistanceOverTime(chaos_particle_A, chaos_particle_B, steps = 500000)
```

```python
sim.plot_double_trajectory(chaos_particle_A, chaos_particle_B, steps = 500000)
```
