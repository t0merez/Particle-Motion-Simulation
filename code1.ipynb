```python
import sim
```

```python
p1 = sim.Particle(position=[0.1, 0, 0], velocity=[0, 100, 0], mass=1e-6, charge=1e-6)
sim.plot_trajectory(p1)
```

```python
orbiting_p = sim.Particle(
    position=[2.0, 0.0, 0.0], 
    velocity=[0.0, 1500.0, 0.0], 
    mass=1.0, 
    charge=-1.0
)
sim.plot_trajectory(orbiting_p)
```

```python
stable_orbit_particle = sim.Particle(
    position=[2.5, 0.0, 0.0], 
    velocity=[0.0, 1800.0, 300.0], # High Y-velocity for orbit, small Z-velocity to bob
    mass=1.0, 
    charge=-1.0
)
sim.plot_trajectory(stable_orbit_particle)
```

```python
figure_8_particle = sim.Particle(
    position=[0.5, 0.0, 0.0], 
    velocity=[0.0, 500.0, 4500.0], # Primary movement is along the Z-axis
    mass=1.0, 
    charge=-1.0
)
sim.plot_trajectory(figure_8_particle)
```

```python
scattering_particle = sim.Particle(
    position=[-5.0, 1.0, 0.5], 
    velocity=[3500.0, 0.0, 0.0], # Fast approach from the left
    mass=1.0, 
    charge=-1.0
)
sim.plot_trajectory(scattering_particle)
```

```python
circular_particle = sim.Particle(
    position=[3.0, 0.0, 0.0], 
    velocity=[0.0, 1450.0, 0.0], # Velocity tuned to this specific radius
    mass=1.0, 
    charge=-1.0
)
sim.plot_trajectory(circular_particle)
```

```python
perfect_circular_particle = sim.Particle(
    position=[2.0, 0.0, 0.0], 
    velocity=[0.0, 0, 0.0],
    mass=1.0, 
    charge=-1.0
)
sim.plot_trajectory(perfect_circular_particle, steps = 5000)
```

```python
centered_particle = sim.Particle(
    position=[0.0, 0.0, 0.0], 
    velocity=[0.0, 0.0, 0.0], 
    mass=1.0, 
    charge=1
)
sim.plot_trajectory(centered_particle, steps = 500)
```
